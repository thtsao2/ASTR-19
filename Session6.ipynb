```python
import math

def get_sin(x):
    return math.sin(x)
```

```python
import math

def get_cos(x):
    return math.cos(x)
```

These two functions above use the math module to return both the cos and sin of x. 

```python
num_entries = 1000
x_values = []
sin_values = []
cos_values = []

start = 0
end = 2
step = (end - start) / (num_entries - 1)

for i in range(num_entries):
    x = start + (i * step)
    x_values.append(x)
    sin_values.append(get_sin(x))
    cos_values.append(get_cos(x))
```

The above cell basically will use the append feature to add each subsequent value of sin(x) and cos(x), going from 0 to 2pi with 1000 values. We also use the two previously defined functions of sin and cos to calculate the value of sin and cos. 

```python
print(f"{'x':>10} | {'sin(x)':>10} | {'cos(x)':>10}")

for i in range(10):
    print(f"{x_values[i]:10.4f} | {sin_values[i]:10.4f} | {cos_values[i]:10.4f}")
```

This cell prints the first 10 values of x, sinx, and cosx neatly on a table, with dashed lines to help differentiate the columns.
